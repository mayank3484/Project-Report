```python
import pandas as pd
```

```python
input_file_path = '/content/drive/MyDrive/AssayingData.xlsx'
```

```python
df=pd.read_excel(input_file_path)
df.head(5)
```

```python
df_reqd = df[['branchname','fedrationname','fponame','locations','procurmentquantitymt','procurmentstatus']]
df_district = df[['branchname','district','fedrationname','locations','procurmentquantitymt','procurmentstatus']]
df_reqd.head(5)
```

```python
df_reqd.columns
```

```python
df_district.head(5)
```

```python
df_district.columns
```

```python
# Filter the data to include only rows with 'approved' or 'pending' procurement status
filtered_df = df_reqd[df_reqd['procurmentstatus'].str.lower().isin(['approved', 'pending'])]
filtered_district_df = df_district[df_district['procurmentstatus'].str.lower().isin(['approved', 'pending'])]
```

```python
filtered_df.columns
```

```python
filtered_df.head(5)
```

```python
# Group by 'branchname', 'federationname', and 'fponame'
grouped_df = filtered_df.groupby(['branchname', 'fedrationname', 'fponame']).agg(
    Total_Location_Count=('locations', 'nunique'),
    Total_Procurement=('procurmentquantitymt', 'sum')).reset_index()
```

```python
group_district_df = filtered_district_df.groupby(['branchname', 'district','fedrationname']).agg(
    Total_Location_Count=('locations', 'nunique'),
    Total_Procurement=('procurmentquantitymt', 'sum')).reset_index()
```

```python
grouped_df.head(5)
```

```python
grouped_df.info()
```

```python
pending_df = df_reqd[df_reqd['procurmentstatus'].str.lower().isin(['pending'])]
pending_district_df = df_district[df_district['procurmentstatus'].str.lower().isin(['pending'])]
pending_df.head(5)
```

```python
pending_df.info()
```

```python
group_pending_df =pending_df.groupby(['branchname', 'fedrationname', 'fponame']).agg(

    Pending=('procurmentquantitymt', 'sum')).reset_index()
```

```python
group_pending_district_df = pending_district_df.groupby(['branchname','district', 'fedrationname']).agg(

    Pending=('procurmentquantitymt', 'sum')).reset_index()
```

```python
group_pending_df.head(5)
```

```python
group_pending_df.info()
```

```python
merged_df = pd.merge(grouped_df, group_pending_df, how='outer', on=list(set(grouped_df.columns).intersection(set(group_pending_df.columns))))
merged_df.head(5)
```

```python
merged_district_df = pd.merge(group_district_df, group_pending_district_df, how='outer', on=list(set(group_district_df.columns).intersection(set(group_pending_district_df.columns))))
merged_district_df.head(5)
```

```python
merged_df =merged_df.fillna(0)
merged_df.head(5)
```

```python
merged_district_df =merged_district_df.fillna(0)
merged_district_df.head(5)
```

```python
# Calculate the pending percentage
merged_df['pending_percentage'] = (merged_df['Pending'] / merged_df['Total_Procurement']) * 100
merged_df.head(5)
```

```python
merged_district_df['pending_percentage'] = (merged_district_df['Pending'] / merged_district_df['Total_Procurement']) * 100
merged_district_df.head(5)
```

```python
merged_df.info()
```

```python
output_file_path = '/content/drive/MyDrive/PendingAssayingReportDated24-07-2024.xlsx'
# Save the output to a new Excel file
merged_df.to_excel(output_file_path, index=False)
```

```python
# Rename columns
merged_df = merged_df.rename(columns={
    'branchname': 'Branch',
    'fedrationname': 'Federation',
    'fponame': 'FPO Name',
    'Total_Location_Count': 'Total Location Count',
    'Total_Procurement': 'Total Procurement(MT)',
    'pending_percentage': '% Pending'
})
```

```python
merged_district_df = merged_district_df.rename(columns={
    'branchname': 'Branch',
    'district': 'District',
    'fedrationname': 'Federation',
    'Total_Location_Count': 'Total Location Count',
    'Total_Procurement': 'Total Procurement(MT)',
    'pending_percentage': '% Pending'
})
```

```python
nafed_federationwise_df = merged_df[(merged_df['Branch']=='NAFED NASIK') |(merged_df['Branch']=='NAFED BHOPAL')]
nafed_federationwise_df.head(5)
```

```python
nafed_districtwise_df = merged_district_df[(merged_district_df['Branch']=='NAFED NASIK') |(merged_district_df['Branch']=='NAFED BHOPAL')]
nafed_districtwise_df.head(5)
```

```python
# Sort the grouped DataFrame by 'pending' in descending order
nafed_federationwise_df = nafed_federationwise_df.sort_values(by='Pending', ascending=False)
nafed_districtwise_df = nafed_districtwise_df.sort_values(by='Pending', ascending=False)
nafed_federationwise_df.head(5)
```

```python
nccf_federationwise_df = merged_df[(merged_df['Branch']=='NCCF NASHIK') |(merged_df['Branch']=='NCCF BHOPAL')]
nccf_federationwise_df.head(5)
```

```python
nccf_districtwise_df = merged_district_df[(merged_district_df['Branch']=='NCCF NASHIK') |(merged_district_df['Branch']=='NCCF BHOPAL')]
nccf_districtwise_df.head(5)
```

```python
# Sort the grouped DataFrame by 'pending' in descending order
nccf_federationwise_df = nccf_federationwise_df.sort_values(by='Pending', ascending=False)
nccf_districtwise_df = nccf_districtwise_df.sort_values(by='Pending', ascending=False)
nccf_federationwise_df.head(5)
```
